```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")
```

```python
file_path = "rbioriginal.xlsx"  # Replace with your file path
sheets = pd.read_excel(file_path, sheet_name=None)  # Read all sheets
df_combined = pd.concat(sheets.values(), ignore_index=True)
```

```python
df_combined.head(15)
```

```python
df = df_combined.iloc[5:].reset_index(drop=True)
df = df.iloc[:, [0, 7, 8]]
df.head(10)
```

```python
#df.columns["date","vol","val"]
df = df.set_axis(["Date","Vol", "Val"], axis=1)#renaming the column 
df.head(10)
```

```python
import datetime 
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df.set_index('Date', inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['Vol'] = df['Vol'].replace('[^\d.]', '', regex=True)
df['Val'] = df['Val'].replace('[^\d.]', '', regex=True)
df[['Vol', 'Val']] = df[['Vol', 'Val']].apply(pd.to_numeric, errors='coerce') 
```
